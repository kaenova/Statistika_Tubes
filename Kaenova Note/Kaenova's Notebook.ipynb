<h3>Importing Data Frame</h3>

```r
df <- read.csv("./BankChurners.csv")
head(df)
descr(df)
```

<h3>Data Wrangling</h3>
Kita hanya akan menggunakan data Customer_Age, Gender, Card Category, Education Level, Martial Status, Income Category, Credit Limit, Total_Trans_Amt, Total_Trans_Ct, dan Avg_Utilization Ratio.   

Satuan:
1. Umur Pengguna dengan satuan tahun  
2. Jenis Kelamin    
    M: Laki-laki    
    F: Perempuan  
3. Tingkat Pendidikan  

{Belum Dilanjut}

```r
#Memindahkan ke dataframe baru
wr <- data.frame("Status_Pelanggan" = df$Attrition_Flag,
                 "Waktu_Buku" = df$Months_on_book,
                 "Umur_Pengguna" = df$Customer_Age,
                 "Jenis_Kelamin" = df$Gender,
                 "Jenis_Kartu" = df$Card_Category,
                 "Tingkat_Pendidikan" = df$Education_Level,
                 "Status_Pernikahan" = df$Marital_Status,
                 "Penghasilan" = df$Income_Category,
                 "Limit_Kredit" = df$Credit_Limit,
                 "Besar_Transaksi" = df$Total_Trans_Amt,
                 "Banyak_Transaksi" = df$Total_Trans_Ct,
                 "Utilisasi" = df$Avg_Utilization_Ratio)
head(wr)
```

<h3>Data Exploration</h3>

Pertama, kita harus tahu keterhubungan antara jenis kartu yang dipakai dan utilisasi kartu tersebut

```r
options(repr.plot.width = 12, repr.plot.height = 6)
par(mfrow=c(1,2))
plot(factor(wr$Jenis_Kartu), wr$Limit_Kredit,
    xlab = "Jenis Kartu",
    ylab = "Limit Kredit")
plot(factor(wr$Jenis_Kartu), wr$Utilisasi,
    xlab = "Jenis Kartu",
    ylab = "Utilisasi Kredit")

par(mfrow=c(1,1))
options(repr.plot.width = 8, repr.plot.height = 8)
pie(table(wr$Status_Pelanggan), 
    labels = paste(names(table(wr$Status_Pelanggan)), "\n", table(wr$Status_Pelanggan), sep=""),
    main = "Pie Chart Status Pelanggan")
```

Dari hasil boxplot diatas, kita bisa melihat bahwa adanya perbedaan yang signifikan dari kartu kredit berjenis Blue yang dengan utilisasi tertinggi dengan limit yang rendah. Yang artinya pengguna yang kurang mengutilisasi kartu kredit tersebut merupakan kartu Gold, Platnium, dan Silver. Pengguna Gold, Silver, dan Platinum sudah diberikan Limit Kredit yang tinggi hanya saja tidak di utilisasikan.  

Utilisasi kredit adalah jumlah kredit yang digunakan / limit kredit.

Sebelum kita memfokuskan kepada kartu-kartu Gold, Platinum dan Silver, kita juga harus mencari tahu bagaimana orang-orang yang memiliki utilisasi yang rendah pada nasabah yang memiliki kartu Blue. Selain hal itu, kita juga akan memfokuskan kepada pengguna yang masih tercatat sebagai pelanggan.

<h4>Blue Card Data Exploration</h4>

```r
blue <- wr[(wr$Jenis_Kartu == "Blue"), ]
blue <- blue[(blue$Status_Pelanggan == "Existing Customer"), ]
head(blue)
descr(blue)
```

```r
plot(factor(blue$Tingkat_Pendidikan), blue$Utilisasi)
```

Dilihat dari boxplot diatas, status pendidikan tidak mempengaruhi utilisasi dari kartu kredit berjenis blue ini. Tetapi pada boxplot ini, terlihat bahwa Collage dan Doctorate memiliki Utilisasi yang rendah terlihat pada Q1 nya yang hingga 0.0.

```r
mean(blue$Waktu_Buku)
cor(blue$Umur_Pengguna, blue$Waktu_Buku)

options(repr.plot.width = 15, repr.plot.height = 12)
par(mfrow=c(2,2))

hist(blue$Umur_Pengguna, breaks = 50)
hist(blue$Waktu_Buku, breaks = 50)

plot(blue$Waktu_Buku, blue$Umur_Pengguna)
hist(blue$Waktu_Buku, breaks = 50, freq = FALSE)
curve(dnorm(x, mean = mean(blue$Waktu_Buku), sd = sd(blue$Waktu_Buku)), add = TRUE, col = "red")
```

Adanya pelonjakan pendaftaran pada ±35 Bulan yang lalu atau sekitar hampir 3 tahun yang lalu, bisa kita lihat dari histogram Waktu Buku. Dapat terlihat juga korelasi terhadap Umur Pengguna dengan Waktu Buku, hanya saja pada sekitaran Waktu Buku 35 bulan tersebut terlihat garis lurus bahwa adanya pelonjakan pendaftaran dari berbagai umur pada waktu tersebut.

```r
options(repr.plot.width = 10, repr.plot.height = 10)
plot_correlation(blue)
```

```r
options(repr.plot.width = 12, repr.plot.height = 8)
plot(blue$Utilisasi, blue$Umur_Pengguna)
```

Dari diagram diatas juga terlihat bahwa kebanyakan orang yang menggunakan utilisasi dibawah rata-rata sekitar umur 40-55.

```r
options(repr.plot.width = 8, repr.plot.height = 6)
plot(factor(blue$Status_Pernikahan), blue$Utilisasi)
```

Kita dapat melihat berapa nasabah yang berada dibawah rata-rata utilisasi pada kategori kartu Blue.

```r
temp <- blue[blue$Utilisasi <= mean(blue$Utilisasi),]

options(repr.plot.width = 18, repr.plot.height = 6)
par(mfrow=c(1,2))
barplot(table(temp$Status_Pernikahan))
barplot(table(temp$Penghasilan))
par(mfrow=c(1,1))
options(repr.plot.width = 10, repr.plot.height = 6)
barplot(table(temp$Tingkat_Pendidikan), space = TRUE)
```

Dapat terlihat, pengguna kartu kategori Blue yang memiliki utilisasi yang rendah dilihat dari status pernikahannya diantara Married (Menikah) dan Single (Belum Menikah). Dari jumlah pendapatan yang terlihat, orang yang memiliki pendapatan di bawah $40K meruapakan tipe orang yang memiliki utilisasi di bawah rata-rata. Setelah itu, orang dengan tingkat pendidikan Graduate merupakan tipe orang yang terbanyak memiliki utilisasi di bawah rata-rata

<h4>Silver, Gold, and Platinum Card Data Exploration</h4>

```r
notblue <- wr[!(wr$Jenis_Kartu == "Blue"), ]
notblue <- notblue[(notblue$Status_Pelanggan == "Existing Customer"), ]
head(notblue)
descr(notblue)
```

```r
options(repr.plot.width = 8, repr.plot.height = 6)
plot(factor(notblue$Status_Pernikahan), notblue$Utilisasi)
```

```r
temp <- notblue[notblue$Utilisasi <= mean(notblue$Utilisasi),]

options(repr.plot.width = 18, repr.plot.height = 6)
par(mfrow=c(1,2))
barplot(table(temp$Status_Pernikahan))
barplot(table(temp$Penghasilan))
par(mfrow=c(1,1))
options(repr.plot.width = 10, repr.plot.height = 6)
barplot(table(temp$Tingkat_Pendidikan), space = TRUE)
```

Mirip dengan pengguna kartu berjenis blue, hanya saja pengguna dengan utilisasi dibawah rata-rata untuk kartu selain biru ini cenderung nasabah yang memiliki penghasilan $60K - $120K
